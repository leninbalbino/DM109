```python
import pandas as pd
import json
from pandas.io.json import json_normalize
import matplotlib.pyplot as pt
df = pd.read_json('telemetry.json', lines=True)
```

```python
# parte 1
df_car3 = json_normalize(df[(df['Car'] == 3)] ['telemetry'])
df_car4 = json_normalize(df[(df['Car'] == 4)] ['telemetry'])
df_car3 = df_car3['Speed'] * (35/9)
df_car4 = df_car4['Speed'] * (35/9)

pt.plot(df_car3, color='red')
pt.plot(df_car4, color='blue')

pt.ylabel('Velocidade (Km/h)')
pt.xlabel('Tempo (s)')
```

```python
# parte 2
df_car3 = pd.read_json('telemetry.json', lines=True)
df_car3 = json_normalize( df_car3[df_car3['Car'] ==3]['telemetry'])
rpm_car3 = df_car3['RPM']
gear_car3 = df_car3['Gear']

f = pt.figure();

ax = f.add_subplot(111)
ax.set_ylabel('Gear',color='blue')
ax.set_xlabel('Time')

pt.plot(gear_car3)

ax2 = ax.twinx()
ax2.plot(rpm_car3, color='green')
ax2.set_ylabel('RPM', color='green')
ax2.tick_params('y', colors='green')

pt.show()
```

```python
# parte 3
df_car3 = pd.read_json('telemetry.json', lines=True)
df_car3 = json_normalize( df_car3[df_car3['Car'] == 3]['telemetry'])
throttle_car3 = df_car3['Throttle']
speed_car3 = df_car3['Speed']  * (35/9)

f = pt.figure();

ax = f.add_subplot(111)
ax.set_ylabel('Speed',color='blue')
ax.tick_params('y', colors='blue')
ax.set_xlabel('Time')

pt.plot(speed_car3)

ax2 = ax.twinx()
ax2.plot(throttle_car3, color='green')
ax2.set_ylabel('Throttle', color='green')
ax2.tick_params('y', colors='green')

pt.show()

# Logo após acelerar o carro, é notado que a velocidade aumenta conforme a intensidade da aceleração e quando a aceleração
# é ausente a velocidade tende a cair.
```
